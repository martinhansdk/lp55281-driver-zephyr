```python
def pwm_val(pwm_in, pwm25, pwm50):
    if brightness < pwm25:
        pwm = pwm_in
        ipls = 0
    elif brightness < pwm50:
        pwm = pwm_in - 32
        ipls = 1
    else:
        pwm = pwm_in - 64
        ipls = 3

    return (ipls << 6) | pwm

result = []
for brightness in range(128):
    value = pwm_val(brightness, pwm25=64, pwm50=96)
    result.append(value)


values == result
```
